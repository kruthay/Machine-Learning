```python
import numpy as np
import pandas as pd
import math
from sklearn.metrics import accuracy_score
```

```python
XTrain = pd.read_csv('train.data', names = ["documentId", "wordId", "wordOccurrence"], sep = ' ')
```

```python
XTest = pd.read_csv('test.data', names = ["documentId", "wordId", "wordOccurrence"], sep = ' ')
m = XTest.wordId.max()
Testgrouped = XTest.groupby(['documentId','wordId']).wordOccurrence.sum()
TestArray = Testgrouped.unstack(level=-1).fillna(0)
TestArray = TestArray.reindex(columns=range(1,m), fill_value=0)
TArray = TestArray.to_numpy()
print(TArray.shape)
```

```python
XTrainGrouped = XTrain.groupby(['documentId','wordId']).wordOccurrence.sum()
TrainArray = XTrainGrouped.unstack(level=-1).fillna(0)
TrainArray = TrainArray.reindex(columns=range(1,m), fill_value=0)
TrainArray.shape
XTrainArray = TrainArray.to_numpy()
print(XTrainArray.shape)
```

```python
Ylabel = pd.read_csv('train.label', names = ['Category'])
```

```python
Ylabel
```

```python
numOFDoc = Ylabel.shape[0]
```

```python
YTrain = Ylabel[Ylabel['Category'] < 21]
```

```python
YTrain['documentId'] = list(range(1, numOFDoc+1))
```

```python
YTrain
```

```python
Ylabel
```

```python
documentinCategory = YTrain.Category.value_counts(sort = False)
```

```python
documentinCategory
```

```python
ProbabilityOfCategory = documentinCategory/numOFDoc
```

```python
ProbabilityOfCategory
```

```python
LogPrCk = np.log(ProbabilityOfCategory)
```

```python
LogPrCk
```

```python
LogPrCkArray = LogPrCk.to_numpy()
```

```python
LogPrCkArray = LogPrCkArray.reshape(1,LogPrCkArray.shape[0])
```

```python
LogPrCkArray.shape
```

```python
DFasWhole = pd.merge(XTrain, YTrain, on='documentId')
```

```python
DFasWhole
```

```python
DFgrouped = DFasWhole.groupby(['Category','wordId']).wordOccurrence.sum()
```

```python
DfAsArray = DFgrouped.unstack(level=1).fillna(0)
```

```python
DfAsArray = DfAsArray.reindex(columns=range(1,m), fill_value=0)
```

```python
LapsmoothDF = DfAsArray+1
```

```python
LapsmoothDF
```

```python
m = DfAsArray.shape[1]
```

```python
m
```

```python
#pwjck = LapsmoothDF.div(LapsmoothDF.sum(axis=1) + m, axis=0)
pwjck = LapsmoothDF.div(LapsmoothDF.sum(axis=1), axis=0)
```

```python
loglikelihood = np.log(pwjck)
```

```python
loglikelihood
```

```python
logLikliehoodarray = loglikelihood.to_numpy()
```

```python
TestArray
```

```python
TArray.shape
```

```python
logLikliehoodarray = np.transpose(logLikliehoodarray)
```

```python
logLikliehoodarray.shape
```

```python
TrainPre = np.matmul(XTrainArray, logLikliehoodarray)
TrainProba = TrainPre + LogPrCkArray
TrainResult = pd.DataFrame(data = TrainProba,index=range(1,11270),columns= range(1,21))
YTrainRes = TrainResult.idxmax(axis=1)
YTrainRes.value_counts()
```

```python
accuracy_score(YTrainRes,Ylabel)
```
